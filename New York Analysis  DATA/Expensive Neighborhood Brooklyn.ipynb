```python
plt.figure(figsize=(20,6))
plt.title('Brooklyn Prices Comparision 2016')
plt.xlabel('Neighbourhood')
plt.ylabel('Prices in Millions')
plt.bar(df2016['NEIGHBORHOOD'],df2016['SALE PRICE'], color = 'b')
plt.xticks(rotation=90)
plt.savefig('Brooklyn Prices Comparision 2016)',bbox_inches='tight')
plt.show()
```
